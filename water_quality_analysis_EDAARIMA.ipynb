# Standardizing the units for each parameter

```python
import pandas as pd
import numpy as np
import re

# ---------- Clean unit strings ----------
def clean_unit(u):
    if pd.isna(u):
        return ''
    u = (
        str(u)
        .strip()
        .lower()
        .replace('°c', 'c')
        .replace('@25 °c', '')
        .replace('@25°c', '')
        .replace('µmhos/cm', 'us/cm')
        .replace('n.t.u.', 'ntu')
        .replace('f.n.u.', 'ntu')
        .replace('ug/l of chl', 'ug/l')
    )
    return u

# Apply to dataframe
param_filtered['uns_name'] = param_filtered['uns_name'].apply(clean_unit)

# ----------  Define conversion logic ----------
def convert_value(val, param, unit):
    """Convert a single numeric value to standardized units."""
    if pd.isna(val):
        return np.nan

    p = str(param).lower().strip()
    u = str(unit).lower().strip()

    # ---- Dissolved Oxygen (% → mg/L) ----
    if 'dissolvedoxygen' in p and '%' in u:
        return (val * 14.6) / 100  # approximate saturation at 25°C, 1 atm

    # ---- Conductance (mS/cm → µS/cm) ----
    elif 'conductance' in p and ('ms/cm' in u):
        return val * 1000

    # ---- Temperature (°F → °C) ----
    elif 'temperature' in p and u in ['f', '°f']:
        return (val - 32) * 5 / 9

    # ---- Secchi Depth (Feet → m; Centimeters → m) ----
    elif 'secchi' in p and 'feet' in u:
        return val * 0.3048
    elif 'secchi' in p and 'centimeter' in u:
        return val * 0.01

    # ---- Flow / Discharge (cfs → m³/s; gallons → m³/s if per sec) ----
    elif (('flow' in p) or ('discharge' in p)) and 'cfs' in u:
        return val * 0.0283168
    elif (('flow' in p) or ('discharge' in p)) and 'gallon' in u:
        return val * 3.78541e-3  # assuming gallons per second

    # ---- Turbidity (NTU/FNU → NTU) ----
    elif 'turbidity' in p and ('ntu' in u or 'fnu' in u):
        return val  # equivalent

    # ---- pH ----
    elif re.search(r'\bph\b', p):
        return val  # already standardized

    # ---- Chlorophyll (RFU → leave unchanged) ----
    elif 'chlorophyll' in p and 'rfu' in u:
        return val  # qualitative unit, keep as-is

    # ---- Everything else already in correct unit ----
    else:
        return val


# ---------- Apply to both numeric columns ----------
def standardize_row(row):
    return pd.Series({
        'fdr_result_std': convert_value(row['fdr_result'], row['parameter'], row['uns_name']),
        'fdr_reporting_limit_std': convert_value(row['fdr_reporting_limit'], row['parameter'], row['uns_name'])
    })

param_filtered[['fdr_result_std', 'fdr_reporting_limit_std']] = param_filtered.apply(standardize_row, axis=1)

# ---------- Assign standardized unit label ----------
def assign_std_unit(param):
    p = param.lower().strip()
    if 'oxygen' in p:
        return 'mg/L'
    elif re.fullmatch(r'ph.*', p) or p.startswith('ph '):
        return 'pH units'
    elif 'temperature' in p:
        return '°C'
    elif 'conductance' in p:
        return 'µS/cm'
    elif 'turbidity' in p:
        return 'NTU'
    elif 'secchi' in p:
        return 'm'
    elif 'flow' in p or 'discharge' in p:
        return 'm³/s'
    elif 'redox' in p:
        return 'mV'
    elif 'chlorophyll' in p:
        return 'µg/L'
    elif 'carbon dioxide' in p:
        return 'mg/L'
    else:
        return None

param_filtered['std_unit'] = param_filtered['parameter'].apply(assign_std_unit)

```

```python
valid_ranges = {
    'DissolvedOxygen_mg/L': (0, 20),
    '(Bottom) DissolvedOxygen_mg/L': (0, 20),
    'Flow, channel_m³/s': (0, 10000),
    'Discharge_m³/s': (0, 10000),
    'Secchi Depth_m': (0, 50),
    'Turbidity_NTU': (0, 1000),
    '(Bottom)Turbidity_NTU': (0, 1000),
    'Turbidity (w/time)_NTU': (0, 1000),
    'SpecificConductance_µS/cm': (0, 50000),
    'ElectricalConductance_µS/cm': (0, 50000),
    '(Bottom) SpecificConductance_µS/cm': (0, 50000),
    'SpecificConductance (EC w/time)_µS/cm': (0, 50000),
    'WaterTemperature_°C': (-2, 50),
    '(Bottom) WaterTemperature_°C': (-2, 50),
    'WaterTemperature (w/time)_°C': (-2, 50),
    'pH_pH units': (0, 14),
    '(Bottom)pH_None': (0, 14),
    'pH (w/time)_pH units': (0, 14),
    'Chlorophyll Volume_µg/L': (0, 1000),
    'Chlorophyll Fluorescence_µg/L': (0, 1000),
    '(Bottom)Chlorophyll Fluorescence_µg/L': (0, 1000),
    'Redox Potential_mV': (-500, 1000),
    'SoilRedox Potential_mV': (-500, 1000),
    'Carbon Dioxide_mg/L': (0, 200)
}
```

```python
def flag_parameter_values(df, param_col='parameter_unit', val_col='fdr_result'):
    """
    Flags parameter values as valid, missing, or out_of_range based on predefined valid_ranges.
    """
    def check_value(param, val):
        if pd.isna(val):
            return 'missing'
        if param not in valid_ranges:
            return 'unknown_param'
        low, high = valid_ranges[param]
        if val < low or val > high:
            return 'out_of_range'
        return 'valid'

    df['value_flag'] = df.apply(lambda x: check_value(x[param_col], x[val_col]), axis=1)
    return df

# Apply to your dataset
cleaned_data = flag_parameter_values(cleaned_data)
```

### Handling the outlier values

```python
summary_before = cleaned_data['value_flag'].value_counts()
print("\nBefore cleaning:")
print(summary_before)

outlier_count = cleaned_data[cleaned_data['value_flag'] == 'out_of_range'].shape[0]

# Set invalid results to NaN
cleaned_data.loc[cleaned_data['value_flag'] == 'out_of_range', 'fdr_result'] = np.nan

# Drop flag column (optional)
cleaned_data = cleaned_data.drop(columns=['value_flag'])

print(f"\nReplaced {outlier_count} out-of-range values with NaN.")


def check_remaining_outliers(df):
    issues = []
    for param, (low, high) in valid_ranges.items():
        mask = df.query("parameter_unit == @param and (fdr_result < @low or fdr_result > @high)")
        if not mask.empty:
            issues.append((param, len(mask)))
    if issues:
        print("\n Remaining outliers detected:")
        for p, c in issues:
            print(f" - {p}: {c} values still out of range")
    else:
        print("\n All parameter values fall within valid environmental ranges.")

check_remaining_outliers(cleaned_data)


param_summary = (
    cleaned_data.groupby('parameter_unit')['fdr_result']
    .agg(['count', 'min', 'max', 'median'])
    .sort_values('count', ascending=False)
)
print("\nSummary of cleaned parameter values:")
print(param_summary.head(10))
```

# Pivoting the table

```python
# Pivot the data so each parameter becomes a column
pivoted = (
    cleaned_data
    .pivot_table(
        index=[
            'station_id', 'station_name', 'station_number', 'full_station_name',
            'station_type', 'latitude', 'longitude', 'county_name',
            'sample_code', 'sample_date', 'year', 'sample_depth_meter'
        ],
        columns='parameter_unit',
        values='fdr_result',
        aggfunc='mean'  # if duplicates exist, take the mean
    )
    .reset_index()
)

# Flatten the multi-index column headers
pivoted.columns.name = None
pivoted.columns = [str(c) for c in pivoted.columns]

print(f"Pivoted dataset shape: {pivoted.shape}")
pivoted.head()
```

Some are bottom-only (prefixed with (Bottom)),

Some are occasional (like Redox Potential, Carbon Dioxide, Flow).

So for each sampling event, only a subset of parameters exists — all others appear as NaN.

```python

# --- Define sub-index (qi) calculation functions ---

def qi_dissolved_oxygen(do):
    # Reference range: 0–14 mg/L; >7 mg/L good
    return np.clip((do / 14) * 100, 0, 100)

def qi_pH(ph):
    # Ideal range 6.5–8.5
    if np.isnan(ph):
        return np.nan
    if 6.5 <= ph <= 8.5:
        return 100
    elif ph < 6.5:
        return 100 - ((6.5 - ph) / 6.5) * 100
    else:  # ph > 8.5
        return 100 - ((ph - 8.5) / 8.5) * 100

def qi_conductivity(cond):
    # Lower is better; <1500 µS/cm = good
    return np.clip(100 - (cond / 1500) * 100, 0, 100)

def qi_turbidity(turb):
    # Lower is better; 0–5 = clear, >100 = poor
    return np.clip(100 - (turb / 100) * 100, 0, 100)

def qi_temperature(temp):
    # Ideal around 20°C; tolerance ±5°C
    return np.clip(100 - abs(temp - 20) * 5, 0, 100)

# --- Compute sub-indices for each parameter ---
pivoted_core['qi_DO'] = pivoted_core['DissolvedOxygen_mg/L'].apply(qi_dissolved_oxygen)
pivoted_core['qi_pH'] = pivoted_core['pH_pH units'].apply(qi_pH)
pivoted_core['qi_Cond'] = pivoted_core['SpecificConductance_µS/cm'].apply(qi_conductivity)
pivoted_core['qi_Turb'] = pivoted_core['Turbidity_NTU'].apply(qi_turbidity)
pivoted_core['qi_Temp'] = pivoted_core['WaterTemperature_°C'].apply(qi_temperature)

# --- Assign parameter weights ---
weights = {'qi_DO': 0.3, 'qi_pH': 0.2, 'qi_Cond': 0.2, 'qi_Turb': 0.2, 'qi_Temp': 0.1}

# --- Calculate WQI (weighted average of available sub-indices) ---
def calc_wqi(row):
    qi_values = []
    qi_weights = []
    for param, weight in weights.items():
        if not np.isnan(row[param]):
            qi_values.append(row[param])
            qi_weights.append(weight)
    if len(qi_values) == 0:
        return np.nan
    return np.average(qi_values, weights=qi_weights[:len(qi_values)])

pivoted_core['WQI'] = pivoted_core.apply(calc_wqi, axis=1)

# --- Classify water quality ---
def classify_wqi(wqi):
    if np.isnan(wqi): return np.nan
    elif wqi >= 80: return 'Good'
    elif wqi >= 50: return 'Moderate'
    elif wqi >= 25: return 'Poor'
    else: return 'Very Poor'

pivoted_core['WQI_Class'] = pivoted_core['WQI'].apply(classify_wqi)
```

```python
# ============================================================
# SMART ARIMA / SARIMA / ARIMAX FORECASTING PIPELINE FOR WQI
# ============================================================
# This version:
#   - auto-detects monthly vs yearly data frequency
#   - supports SARIMA (seasonal) when monthly data available
#   - can include exogenous predictors (Conductance, Temp, etc.)
#   - filters unstable series and prints clear diagnostics
# ============================================================

from statsmodels.tsa.statespace.sarimax import SARIMAX
from itertools import product
import warnings
warnings.filterwarnings("ignore", message="Maximum Likelihood optimization failed")

# ------------------------------------------------------------
# CONFIGURATION
# ------------------------------------------------------------
FORECAST_TARGET = "WQI"         # can be 'WQI' or a specific parameter
HORIZON_STEPS = 12              # 12 months ahead (or 3 years if annual)
USE_EXOGENOUS = True            # include other parameters as predictors
MIN_SERIES_LENGTH = 18          # min data points required
MAX_MISSING_RATIO = 0.3         # skip if >30% missing

# ------------------------------------------------------------
# UTILITY: detect frequency of your dataset
# ------------------------------------------------------------
def detect_freq(series_index):
    if len(series_index) < 3:
        return "Y"
    diffs = np.median(np.diff(series_index.astype(int)))
    return "M" if diffs < 3 else "Y"

# ------------------------------------------------------------
# FIT BEST MODEL (ARIMA, SARIMA, ARIMAX)
# ------------------------------------------------------------
def fit_best_model(ts, exog=None, freq="Y"):
    """
    Chooses best (p,d,q) and seasonal order if monthly.
    """
    best_aic = np.inf
    best_model, best_order, best_seasonal = None, None, None

    # Candidate grids
    pdq_grid = list(product(range(0, 3), range(0, 2), range(0, 3)))
    seasonal_grid = [(0, 1, 1, 12), (1, 1, 1, 12)] if freq == "M" else [(0, 0, 0, 0)]

    for (p, d, q) in pdq_grid:
        for (P, D, Q, s) in seasonal_grid:
            try:
                model = SARIMAX(
                    ts,
                    exog=exog,
                    order=(p, d, q),
                    seasonal_order=(P, D, Q, s),
                    trend="n",
                    enforce_stationarity=False,
                    enforce_invertibility=False
                ).fit(disp=False, maxiter=3000, tol=1e-6)

                if model.mle_retvals.get("converged", False) and model.aic < best_aic:
                    best_aic, best_model = model.aic, model
                    best_order, best_seasonal = (p, d, q), (P, D, Q, s)
            except Exception:
                continue

    return best_model, best_order, best_seasonal

# ------------------------------------------------------------
# BACKTEST + FORECAST FUNCTION
# ------------------------------------------------------------
def backtest_and_forecast(ts, exog=None, horizon=HORIZON_STEPS):
    ts = ts.sort_index()
    freq = detect_freq(ts.index)
    if len(ts.dropna()) < MIN_SERIES_LENGTH or ts.std() < 1e-3:
        return None

    test_len = min(max(3, len(ts)//10), 12)
    train, test = ts.iloc[:-test_len], ts.iloc[-test_len:]
    exog_train, exog_test = (None, None)
    if exog is not None:
        exog_train, exog_test = exog.iloc[:-test_len], exog.iloc[-test_len:]

    model, order, seasonal = fit_best_model(train, exog_train, freq)
    if model is None:
        return None

    preds = model.get_forecast(steps=test_len, exog=exog_test).predicted_mean
    mae = float(np.mean(np.abs(test.values - preds.values)))
    mape = float(np.mean(np.abs((test.values - preds.values) /
                                np.where(test.values == 0, 1e-9, test.values))) * 100)

    # Refit full series for forecasting
    full_model, full_order, full_seasonal = fit_best_model(ts, exog, freq)
    if full_model is None:
        return None

    # Prepare future exogenous values (if exog used)
    future_exog = None
    if exog is not None:
    # Repeat the last available row for horizon steps
        last_row = exog.iloc[[-1]].values
        future_exog = np.repeat(last_row, horizon, axis=0)
        future_exog = pd.DataFrame(future_exog, columns=exog.columns)

# Forecast future steps
    fc = full_model.get_forecast(steps=horizon, exog=future_exog).predicted_mean
    fc.index = [ts.index.max() + i for i in range(1, horizon + 1)]

    fc.index = [ts.index.max() + i for i in range(1, horizon + 1)]

    return {
        "freq": freq,
        "order": full_order,
        "seasonal": full_seasonal,
        "mae": mae,
        "mape": mape,
        "forecast": fc
    }


# ============================================================
# Step 7: Run ARIMA/SARIMA/ARIMAX models per station
# ============================================================

forecast_rows = []

for sid, grp in county_df.groupby("station_id"):
    # --------------------------------------------------------
    # Select the correct column based on target
    # --------------------------------------------------------
    if FORECAST_TARGET == "WQI":
        target_col = "WQI_mean"
    else:
        target_col = f"{FORECAST_TARGET}_mean"
    if target_col not in grp.columns:
        continue

    # --------------------------------------------------------
    # Aggregate to ensure one observation per period (e.g., per year)
    # --------------------------------------------------------
    series = (
        grp.groupby("year")[target_col]
           .mean()
           .sort_index()
           .dropna()
    )

    # Skip too-short or constant series early
    if len(series) < MIN_SERIES_LENGTH or series.std() < 1e-3:
        continue

    # --------------------------------------------------------
    # Prepare consistent PeriodIndex and fill missing years
    # --------------------------------------------------------
    series.index = pd.PeriodIndex(series.index, freq="Y")
    series = series.asfreq("Y")
    series = series.interpolate(limit_direction="both")

    if series.isna().mean() > MAX_MISSING_RATIO:
        continue  # skip if too many missing years

    # --------------------------------------------------------
    # Prepare exogenous predictors (ARIMAX)
    # --------------------------------------------------------
    exog = None
    if USE_EXOGENOUS:
        possible_exog = [
            "SpecificConductance_µS/cm_mean",
            "Turbidity_NTU_mean",
            "WaterTemperature_°C_mean",
            "DissolvedOxygen_mg/L_mean"
        ]
        valid_exog = [c for c in possible_exog if c in grp.columns]

        if valid_exog:
            exog = (
                grp.groupby("year")[valid_exog]
                   .mean()
                   .sort_index()
            )
            exog.index = pd.PeriodIndex(exog.index, freq="Y")
            exog = exog.asfreq("Y").interpolate(limit_direction="both")

    # --------------------------------------------------------
    # Fit model and forecast
    # --------------------------------------------------------
    result = backtest_and_forecast(series, exog=exog, horizon=HORIZON_STEPS)
    if result is None:
        continue

    # --------------------------------------------------------
    # Collect forecasts for output
    # --------------------------------------------------------
    for year, val in result["forecast"].items():
        forecast_rows.append({
            "station_id": sid,
            "county_name": grp["county_name"].iloc[0],
            "latitude":    grp["latitude"].iloc[0],
            "longitude":   grp["longitude"].iloc[0],
            "target":      FORECAST_TARGET,
            "order":       result["order"],
            "seasonal_order": result["seasonal"],
            "mae":         result["mae"],
            "mape":        result["mape"],
            "forecast_period": str(year),
            "forecast_value": float(val),
            "last_observed_year": int(series.index.max().year)
        })

# ============================================================
# Step 8: Validation & Output Summary
# ============================================================
forecast_df = pd.DataFrame(forecast_rows)

print("\nForecast value distribution:")
print(forecast_df["forecast_value"].describe())

print("\nMAPE distribution:")
print(forecast_df["mape"].describe())

# Compare last observed vs forecast values
validation = (
    county_df.groupby("station_id")[target_col]
             .last()
             .reset_index()
)
merged = forecast_df.merge(validation, on="station_id", how="left")

merged["change_%"] = np.where(
    merged[target_col] == 0,
    np.nan,
    100 * (merged["forecast_value"] - merged[target_col]) / merged[target_col]
)

print("\nPercentage change between last observed and forecast values:")
print(merged["change_%"].describe())

# Save results
output_path = f"forecasts_{FOCAL_COUNTY}_{FORECAST_TARGET}_smart.csv"
forecast_df.to_csv(output_path, index=False)
print(f"\nEnhanced forecasts saved → {output_path}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
import numpy as np

def rolling_forecast_validation(series, p=0, d=1, q=1, window=10, horizon=1):
    """
    Perform rolling-origin (walk-forward) validation for ARIMA.
    - series: pandas Series with PeriodIndex
    - (p,d,q): ARIMA order
    - window: number of training observations before starting validation
    - horizon: number of steps to forecast each time (default = 1 year ahead)
    """
    from statsmodels.tsa.statespace.sarimax import SARIMAX
    errors_mae, errors_mape = [], []

    for i in range(window, len(series) - horizon):
        train = series.iloc[:i]
        test = series.iloc[i:i + horizon]

        try:
            model = SARIMAX(train, order=(p, d, q), trend='n',
                            enforce_stationarity=False,
                            enforce_invertibility=False).fit(disp=False)
            preds = model.get_forecast(steps=horizon).predicted_mean
            mae = mean_absolute_error(test, preds)
            mape = mean_absolute_percentage_error(test, preds) * 100
            errors_mae.append(mae)
            errors_mape.append(mape)
        except:
            continue

    return np.mean(errors_mae), np.std(errors_mae), np.mean(errors_mape), np.std(errors_mape)
```

```python
# Aggregate across all stations in a county per year
county_series = (
    county_df.groupby(['county_name','year'])
             [['WQI_mean','DissolvedOxygen_mg/L_mean',
               'SpecificConductance_µS/cm_mean','Turbidity_NTU_mean',
               'WaterTemperature_°C_mean','pH_pH units_mean']]
             .mean()
             .reset_index()
)

# Example: San Joaquin County
sj = county_series[county_series['county_name']=='San Joaquin'].set_index('year')

# Example: ADF test on county-level WQI
from statsmodels.tsa.stattools import adfuller
result = adfuller(sj['WQI_mean'].dropna())
print(f"ADF Statistic: {result[0]:.3f}, p-value: {result[1]:.3f}")
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# 1. Prepare series
ts = sj['WQI_mean'].dropna()

# 2. Check stationarity
from statsmodels.tsa.stattools import adfuller
adf = adfuller(ts)
print(f"ADF Statistic: {adf[0]:.3f}, p-value: {adf[1]:.3f}")

# 3. Difference if non-stationary (p>0.05)
if adf[1] > 0.05:
    ts = ts.diff().dropna()

# 4. Fit ARIMA at county level
model = SARIMAX(ts, order=(0,1,2)).fit(disp=False)
print(model.summary())

# 5. Forecast next 3 years
forecast = model.get_forecast(steps=3)
print(forecast.predicted_mean)
```

Model Validation (San Joaquin County):
The ADF test (ADF = −1.70, p = 0.43) confirmed that the WQI time series was non-stationary.
ARIMA(0,1,2) was therefore selected to difference the data once and model the remaining autocorrelation.
Residual diagnostics (Ljung–Box p = 0.62, JB p = 0.72) indicated no autocorrelation and approximately normal errors, confirming model adequacy.
The model captured temporal dynamics effectively, with forecasts suggesting stable water quality conditions over the next few years.

```python
# --- Step 1: Invert differencing to restore original scale ---
forecast_values = forecast.predicted_mean.cumsum() + ts.iloc[-1]

# --- Step 2: Generate future year index based on your observed years ---
# If your ts index is in years (like 2000, 2001, ..., 2025)
last_year = int(ts.index[-1])                       # last observed year
forecast_years = list(range(last_year + 1, last_year + len(forecast_values) + 1))

# Create a new series for forecast with proper year index
forecast_series = pd.Series(forecast_values.values, index=forecast_years)

# --- Step 3: Plot observed + forecast together on the same axis ---
plt.figure(figsize=(10,4))
plt.plot(ts.index, ts.values, marker='o', label='Observed', color='blue')
plt.plot(forecast_series.index, forecast_series.values, marker='o', color='red', label='Forecast')
plt.title("Observed vs Forecasted WQI")
plt.xlabel("Year"); plt.ylabel("WQI")
plt.legend(); plt.grid(True)
plt.show()
```
